```python
# The mean by the type of University
mean_type = df.groupby('Type')
mean_type = mean_type.mean().sort_values(by='Size',ascending=True)
mean_type
```

```python
add = df['address']
add = add.str.strip()
add = add.str.rsplit(',',2,expand=True)
add
```

```python
df['state_zip']= add[2]
df
```

```python
df['city']= add[1]
df
```

```python
new2 = df["state_zip"].str.split(" ",n = 2 ,expand = True)
df['state'] = new2[1]
df['zipcode'] = new2[2]
df
```

```python
df = df.drop(columns=[ 'state_zip','index'])
```

```python
df
```

```python
associates = []
for i in df['Type']:
    if 'primarily' in i:
        associates.append(1)
    else:
        associates.append(0)
```

```python
df["associates"] = associates
```

```python
df
```

```python
Profit_or_non = []
for i in df['Type']:
    if 'not-for' in i:
        Profit_or_non.append('Not-For-Profit')
    elif  'Public' in i:
        Profit_or_non.append('Not-For-Profit')
    else:
        Profit_or_non.append('Profit')
```

```python
df["Profit"] = Profit_or_non
```

```python
df.head(30)
```

```python
Sector = []
for i in df['Type']:
    if 'Pulic' in i:
        Sector.append('Public')
    else:
        Sector.append('Private')
```

```python
df['Sector'] = Sector
df
```
